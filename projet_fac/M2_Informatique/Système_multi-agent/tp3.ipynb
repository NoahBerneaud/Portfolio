```python
import networkx as nx

class Piece:
    def __init__(self, name, liste_pieces_suivantes, tableau, torche, fenetre, tresor):
        self.name = name
        self.liste_pieces_suivantes = liste_pieces_suivantes
        self.tableau = tableau
        self.torche = torche
        self.fenetre = fenetre
        self.tresor = tresor
    
    def __str__(self):
        return "Piece " + str(self.name) 
    
    def contains_tresor(self):
        return self.tresor
    
    def contains_fenetre(self):
        return self.fenetre
    
    def contains_torche(self):
        return self.torche
    
    def set_liste_pieces_suivantes(self, liste_pieces_suivantes):
        self.liste_pieces_suivantes = liste_pieces_suivantes
    
    def get_liste_pieces_suivantes(self):
        return self.liste_pieces_suivantes
    
    def constrains_tableau(self):
        return self.tableau
    
    def contains_objet(self, list_objets):
        for objet in list_objets:
            if objet == "tableau":
                if not self.constrains_tableau():
                    return False
            elif objet == "torche":
                if not self.contains_torche():
                    return False
            elif objet == "fenetre":
                if not self.contains_fenetre():
                    return False
            elif objet == "tresor":
                if not self.contains_tresor():
                    return False
        return True
        
    
    def find_Objets(self,historique=set(),list_objets=[]):
        
        print("Je suis dans la pièce : ", self.name)
        if self.contains_objet(list_objets):
            return self.name
        else:
            for piece in self.get_liste_pieces_suivantes():
                if piece not in historique:
                    historique.add(piece)
                    solution = piece.find_Objets(historique, list_objets)
                    if(solution != None):
                        return solution
                    historique.remove(piece)
        return None
    
    def chemin_avec_objets(self, list_objets=[], historique=set()):
        if not self.contains_objet(list_objets):
            return None
        else:
            historique.add(self)
            for piece in self.get_liste_pieces_suivantes():
                if(piece in historique):
                    return historique
                else:
                    solution = piece.chemin_avec_objets(list_objets, historique)
                    if(solution != None):
                        return solution
                    
            historique.remove(self)
    
    def chemin_strict_avec_objets(self, list_objets=[], historique=set()):
        if not self.contains_objet(list_objets):
            return None
        else:
            historique.add(self)
            for piece in self.get_liste_pieces_suivantes():
                if(piece in historique):
                    return historique
                else:
                    solution = piece.chemin_strict_avec_objets(list_objets, historique)
                    if(solution is None):
                        return None
            return historique
    
    def question_17(self):
        #Trouver une pièce avec un tableau
        #Regarder les voisins n+2 et voir s'ils ont un trésor 
        #Si oui, on a trouvé une solution
        if(self.constrains_tableau()):
            pièce_tableau = self
        else: 
            return None 
        while(pièce_tableau is None):
            for piece in self.get_liste_pieces_suivantes():
                if(piece.constrains_tableau()):
                    pièce_tableau = piece

        for piece in pièce_tableau.get_liste_pieces_suivantes():
            for piece2 in piece.get_liste_pieces_suivantes():
                if piece2.contains_tresor():
                    return True
        return False
                   
        
    
            
                    
                
            
        
        


piece_1 = Piece(1, set(), tableau=True, torche=True, fenetre=False, tresor=False)

piece_2 = Piece(2, set(), tableau=True, torche=False, fenetre=False, tresor=False)

piece_3 = Piece(3, set(), tableau=False, torche=True, fenetre=False, tresor=True)

piece_4 = Piece(4, set(), tableau=False, torche=True, fenetre=True, tresor=False)

piece_5 = Piece(5, set(), tableau=True, torche=True, fenetre=False, tresor=False)

piece_6 = Piece(6, set(), tableau=True, torche=False, fenetre=False, tresor=False)

piece_7 = Piece(7,  set(), tableau=False, torche=False, fenetre=False, tresor=True)

piece_8 = Piece(8, set(), tableau=False, torche=False, fenetre=True, tresor=False)

piece_9 = Piece(9, set(), tableau=True, torche=False, fenetre=True, tresor=False)

piece_10 = Piece(10, set(), tableau=False, torche=True, fenetre=False, tresor=False)

piece_1.set_liste_pieces_suivantes(set([piece_2, piece_5]))
piece_2.set_liste_pieces_suivantes(set([piece_3, piece_1]))
piece_3.set_liste_pieces_suivantes(set([piece_3, piece_4]))
piece_4.set_liste_pieces_suivantes(set([piece_2, piece_6, piece_8, piece_10]))
piece_5.set_liste_pieces_suivantes(set([piece_5, piece_4]))
piece_6.set_liste_pieces_suivantes(set([piece_5,piece_3]))
piece_7.set_liste_pieces_suivantes(set([piece_5, piece_6, piece_8]))
piece_8.set_liste_pieces_suivantes(set([piece_6, piece_9, piece_3]))
piece_9.set_liste_pieces_suivantes(set([piece_10]))
piece_10.set_liste_pieces_suivantes(set([piece_9]))

print(piece_2.question_17())
```
