```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import sklearn as sk
```

```python
# Importing the dataset
x_train = np.load('regression_xtrain.npy')
y_train = np.load('regression_ytrain.npy')
x_test = np.load('regression_xtest.npy')
y_test = np.load('regression_ytest.npy')
```

```python
# Visualising the Training set results
plt.scatter(x_train, y_train, color = 'red')
# Visualising the Test set results
plt.scatter(x_test, y_test, color = 'blue')
```

```python
# Fitting Linear Regression to the dataset
from sklearn.linear_model import LinearRegression
x_train = x_train.reshape(-1,1)
y_train = y_train.reshape(-1,1)
x_test = x_test.reshape(-1,1)
y_test = y_test.reshape(-1,1)
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
```

```python
print("Score train: ", lin_reg.score(x_train, y_train))
print("Score test: ", lin_reg.score(x_test, y_test))
```

```python
x_d = np.linspace(0, 1, 100)
plt.scatter(x_train, y_train, color = 'blue')
plt.scatter(x_test, y_test, color = 'orange')
plt.plot(x_d, lin_reg.coef_[0]*x_d+lin_reg.intercept_, color = 'red')
```

```python
plt.hist(x_test, bins=20)
hist = np.histogram(x_test, bins=20)
print(hist[0])
```

```python
plt.hist(x_train, bins=20)
tab = np.argmin(x_train>hist[1],axis=1)
print(hist[0][tab])
```

```python
newLinReg = LinearRegression()
newLinReg.fit(x_train, y_train, sample_weight=hist[0][tab])
print("Score train: ", newLinReg.score(x_train, y_train))
print("Score test: ", newLinReg.score(x_test, y_test))
```

```python
plt.scatter(x_train, y_train, color = 'blue')
plt.scatter(x_test, y_test, color = 'orange')
plt.plot(x_d, newLinReg.coef_[0]*x_d+newLinReg.intercept_, color = 'red')
```
